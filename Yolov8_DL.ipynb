<p style="background-color:;font-family:newtimeroman;color:#D5CABD;font-size:120%;text-align:center;border-radius:20px 60px;">‚ÄúA Review of Face Anti-spoofing Methods for Face Recognition Systems‚Äù DOI:10.1109/INISTA52262.2021.9548404Conference: 2021 International Conference on Innovations in Intelligent SysTemsand Applications (INISTA)</p> 

#### An ordinary face recognition system lacks concrete anti-spoofing measures, making it vulnerable to spoofing attacks where a registered user's photos or videos can be easily obtained and used without consent. As the popularity of face recognition systems increases, so does the need for anti-spoofing measures to counter such attacks. Researchers have devised various methods for face anti-spoofing, including physiological cues like blinking, head rotation, or facial expressions, as well as texture analysis, image quality analysis, and image distortion analysis for spoof detection.
#### Efforts to develop face anti-spoofing methods have led to the availability of publicly accessible face anti-spoofing databases, easing researchers' access to data for developing efficient anti-spoofing techniques. These databases have encouraged advancements in countermeasures against face spoofing by focusing on the design of effective anti-spoof methods. Researchers have proposed active methods that require user interaction for liveness detection and passive methods that do not need user participation, making the system non-intrusive for users.
#### Face spoof attacks come in 2D and 3D forms, with 2D attacks like print/photo attacks involving printed or screen-displayed photos, cut-photo attacks, or video attacks that loop a user's face. In contrast, 3D attacks use 3D masks, makeup, or plastic surgeryto deceive face recognition systems, making it harder to detect such attacks. Various classification methods based on liveness cues, such as motion, texture, image quality, depth information, multimodal approaches, ensemble descriptors, deep learning methods, and sensor-level approaches, have been explored to detect face spoofing.Researchers have proposed methods like enhanced texture-based approaches, depth-supervised learning, Multi-Regional Convolutional Neural Networks (MRCNN), and two-stream CNN architectures for face anti-spoofing, achieving impressive results in detecting spoofed faces in different datasets. By combining features from multiple descriptors, utilizing depth information, and employing advanced neural network architectures, researchers aim to enhance the robustness and efficiency of face anti-spoofing methods to combat evolving spoofing techniques effectively.The taxonomy of face anti-spoofing methods categorizes them into active and passive approaches based on user participation, while classification based on liveness cues distinguishes methods using motion, texture, or depth information. Research methods are selected through a systematic study selection procedure, enhancing the understanding and detection of face spoofing attacks through cutting-edge anti-spoofing techniques and advancements in the field.

RealCollection

```python
classID = 1                                    #clasifire               # 0 is fake and 1 is real
outputFolderPath = 'Dataset/real'
confidence = 0.8
save = True
blurThreshold = 25  
num_augmentations = 4 
debug = False
offsetPercentageW = 10
offsetPercentageH = 20
camWidth, camHeight = 640, 480
floatingPoint = 6
####################################

os.environ['CUDA_VISIBLE_DEVICES'] = '0'

cap = cv2.VideoCapture(0)
cap.set(3, camWidth)
cap.set(4, camHeight)

############### augmentation function ####################
detector = FaceDetector()
                                                  
def augment_imag(image):
                                                  #  random augmentation techniques here
    aug_ima = image.copy()
    
                                                  # flip horizontally with a probability of 0.5
    if np.random.rand() < 0.5:
        aug_ima = cv2.flip(aug_ima, 1)  # flip horizontally
    
                                                  #  rotation between -30 to 30 degrees
    angle = np.random.randint(-30, 30)
    rows, cols, _ = aug_ima.shape
    rotation_matrix = cv2.getRotationMatrix2D((cols/2, rows/2), angle, 1)
    aug_ima = cv2.warpAffine(aug_ima, rotation_matrix, (cols, rows))
    
                                                    # scaling between 0.5 to 1.2
    scale = np.random.uniform(0.5, 1.2)
    aug_ima = cv2.resize(aug_ima, None, fx=scale, fy=scale, interpolation=cv2.INTER_LINEAR)
    
                                                     #  brightness adjustment
    bright_fac = np.random.uniform(0.5, 1.5)  
    aug_ima = cv2.convertScaleAbs(aug_ima, alpha=bright_fac, beta=0)
    
    return aug_ima

while True:
    success, img = cap.read()
    imgOut = img.copy()
    img, bboxs = detector.findFaces(img, draw=False)

    listBlur = []                                       # true false values indicating if the faces are blur or not
    listInfo = []                                       # the normalized values and the class name for the label txt file
    if bboxs:
################ bboxInfo - "id","bbox","score","center"
        for bbox in bboxs:
            x, y, w, h = bbox["bbox"]
            score = bbox["score"][0]

################ check the score
            if score > confidence:
                                                             # adding an offset to the face Detected
                offsetW = (offsetPercentageW / 100) * w
                x = int(x - offsetW)
                w = int(w + offsetW * 2)
                offsetH = (offsetPercentageH / 100) * h
                y = int(y - offsetH * 3)
                h = int(h + offsetH * 3.5)

                # to avoid values below 0
                x = max(0, x)
                y = max(0, y)
                w = max(0, w)
                h = max(0, h)

################ find Blurriness
                imgFace = img[y:y + h, x:x + w]
                blurValue = int(cv2.Laplacian(imgFace, cv2.CV_64F).var())
                if blurValue > blurThreshold:
                    listBlur.append(True)
                else:
                    listBlur.append(False)

############### normalize Values
                ih, iw, _ = img.shape
                xc, yc = x + w / 2, y + h / 2

                xcn, ycn = round(xc / iw, floatingPoint), round(yc / ih, floatingPoint)
                wn, hn = round(w / iw, floatingPoint), round(h / ih, floatingPoint)

                xcn = min(1, xcn)
                ycn = min(1, ycn)
                wn = min(1, wn)
                hn = min(1, hn)

                listInfo.append(f"{classID} {xcn} {ycn} {wn} {hn}\n")

############### drawing
                cv2.rectangle(imgOut, (x, y, w, h), (255, 0, 0), 3)
                cvzone.putTextRect(imgOut, f'Score: {int(score * 100)}% Blur: {blurValue}', (x, y - 0),
                                   scale=2, thickness=3)
                if debug:
                    cv2.rectangle(img, (x, y, w, h), (255, 0, 0), 3)
                    cvzone.putTextRect(img, f'Score: {int(score * 100)}% Blur: {blurValue}', (x, y - 0),
                                       scale=2, thickness=3)

##############  Save
        if save:
            if all(listBlur) and listBlur != []:
                                                                           # save augmented images
                for _ in range(num_augmentations):  
                    augmented_img = augment_imag(img)                       # apply augmentation
                    timeNow = time()                                         # Get current time
                    timeNow = str(timeNow).split('.')
                    timeNow = timeNow[0] + timeNow[1]
                                                                            # save augmented image
                    cv2.imwrite(f"{outputFolderPath}/{timeNow}.jpg", augmented_img)
                                                                                # save corresponding label file (if needed)
                    for info in listInfo:
                        f = open(f"{outputFolderPath}/{timeNow}.txt", 'a')
                        f.write(info)
                        f.close()

    cv2.imshow("Image", imgOut)
    cv2.waitKey(1)
```

```python
import os
import random
import shutil
from itertools import islice

outputFolderPath = "Dataset/SplitData"
inputFolderPath = "Dataset/all"
splitRatio = {"train": 0.6, "val": 0.2, "test": 0.2}
classes = ["fake","real"]

try:
    shutil.rmtree(outputFolderPath)
except OSError as e:
    os.mkdir(outputFolderPath)

# --------  Directories to Create -----------
os.makedirs(f"{outputFolderPath}/train/images", exist_ok=True)
os.makedirs(f"{outputFolderPath}/train/labels", exist_ok=True)
os.makedirs(f"{outputFolderPath}/val/images", exist_ok=True)
os.makedirs(f"{outputFolderPath}/val/labels", exist_ok=True)
os.makedirs(f"{outputFolderPath}/test/images", exist_ok=True)
os.makedirs(f"{outputFolderPath}/test/labels", exist_ok=True)

# --------  Get the Names  -----------
listNames = os.listdir(inputFolderPath)

uniqueNames = []
for name in listNames:
    uniqueNames.append(name.split('.')[0])
uniqueNames = list(set(uniqueNames))

# --------  Shuffle -----------
random.shuffle(uniqueNames)

# --------  Find the number of images for each folder -----------
lenData = len(uniqueNames)
lenTrain = int(lenData * splitRatio['train'])
lenVal = int(lenData * splitRatio['val'])
lenTest = int(lenData * splitRatio['test'])

# --------  Put remaining images in Training -----------
if lenData != lenTrain + lenTest + lenVal:
    remaining = lenData - (lenTrain + lenTest + lenVal)
    lenTrain += remaining

# --------  Split the list -----------
lengthToSplit = [lenTrain, lenVal, lenTest]
Input = iter(uniqueNames)
Output = [list(islice(Input, elem)) for elem in lengthToSplit]
print(f'Total Images:{lenData} \nSplit: {len(Output[0])} {len(Output[1])} {len(Output[2])}')

# --------  Copy the files  -----------

sequence = ['train', 'val', 'test']
for i,out in enumerate(Output):
    for fileName in out:
        shutil.copy(f'{inputFolderPath}/{fileName}.jpg', f'{outputFolderPath}/{sequence[i]}/images/{fileName}.jpg')
        shutil.copy(f'{inputFolderPath}/{fileName}.txt', f'{outputFolderPath}/{sequence[i]}/labels/{fileName}.txt')

print("Split Process Completed...")


# -------- Creating Data.yaml file  -----------

dataYaml = f'path: ../Data\n\
train: ../train/images\n\
val: ../val/images\n\
test: ../test/images\n\
\n\
nc: {len(classes)}\n\
names: {classes}'



f = open(f"{outputFolderPath}/data.yaml", 'a')
f.write(dataYaml)
f.close()

```

The top1_acc and top5_acc metrics, which show the accuracy of the model on the validation set, are used to evaluate the performance and accuracy of the trained model. During training, these measurements are given after each epoch.

How to interpret metrics
Top-1 accuracy (top1_acc), this statistic shows the true label and the one-match ratio of the model's best prediction‚Äîthe highest confidence score. In the example top1_acc 0.5 indicates that the model predicts the label 50% correctly.

Top-5 accuracy (top5_acc): This measure is often used in multi-class classification problems, where the true label is considered correct if it appears in the top 5 predictions made by the model. With two classes, the top5_ac measure is not really useful for binary classification assignment.

Performance Evaluation Top-1 Accuracy 50%: When a model achieves a top1_acc of 50% for a binary classification task, it is performing at the random guess level. For a binary classification task, this is a poor performance statistic, since a stronger model should produce greater accuracy.
Increasing the accuracy of the model
Here are some ideas to improve the model if the accuracy is not the same:

Data augmentation: Use augmentation strategies to make the training data more diverse.
Hyperparameter Tuning: Play with different hyperparameters such as batch size, learning rate and optimization settings.
Model architecture: Consider using a more complex or different model architecture.
More data: Adding more training data can greatly improve the performance of the model.
Transfer learning: Develop a pre-trained model, using your specific data set.

```python
import math
import time
import cv2
import cvzone
from ultralytics import YOLO
import os
import numpy as np

confidence_yolo = 0.6

yolo_model = YOLO("../models/best.pt")
class_names_yolo = ["fake", "real"]

cap = cv2.VideoCapture(0)
cap.set(3, 1280)
cap.set(4, 720)

while True:
    success, img = cap.read()

    # (real and fake detection)
    results_yolo = yolo_model(img, stream=True, verbose=False)

    if results_yolo:
        for r in results_yolo:
            boxes = r.boxes
            if boxes:
                for box in boxes:
                    x1, y1, x2, y2 = map(int, box.xyxy[0])
                    w, h = x2 - x1, y2 - y1
                    conf = math.ceil((box.conf[0] * 100)) / 100
                    cls = int(box.cls[0])

                    if conf > confidence_yolo:
                        if class_names_yolo[cls] == 'real':
                            color = (255, 0, 0)
                        else:
                            color = (0, 0, 255)

                        cvzone.cornerRect(img, (x1, y1, w, h), colorC=color, colorR=color)
                        cvzone.putTextRect(img, f'{class_names_yolo[cls].upper()} {int(conf*100)}%',
                                           (max(0, x1), max(35, y1)), scale=2, thickness=4, colorR=color, colorB=color)

   
    cv2.imshow('Image', img)

   
    if cv2.waitKey(1) == ord('q'):
        break


cap.release()
cv2.destroyAllWindows()
```
